# EDA ==> Exploratory Data Analysis 

# Parts of EDA
1. Univariate Analysis ==> Analysis on a single Independent Column
2. Bivariate Analysis ==> Analysis on two Columns
3. Multivariate Analysis ==> Analysis on more than 2 coloumns
